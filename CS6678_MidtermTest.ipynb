```python
#1 Covid-19 Cases: Calculate the death rates of all counties in AL, sort the counties by the death rate in descending order.
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# read csv file
df = pd.read_csv("COVID-19 in Alabama-2.csv")
# change data to numeric type
df['Cases'] = pd.to_numeric(df['Cases'], errors='coerce')
df['Deaths'] = pd.to_numeric(df['Deaths'], errors='coerce')
total_cases = df.groupby('Counties')['Cases'].sum()
total_deaths = df.groupby('Counties')['Deaths'].sum()
death_rate = (total_deaths/total_cases).sort_values(ascending = False)
print(death_rate)
plt.figure(figsize=(5, 15))
sns.set_color_codes("pastel")
sns.barplot(y = death_rate.index, x = death_rate.values, orient = "h", color="b")

plt.title('Death Rate (descending order) ')
plt.xlabel('Death Rate: Deaths / Cases')
plt.ylabel('County')
plt.show()


```

```python
#2 Covid-19 Vaccine Providers: Group all the vaccine providers in the same county in AL, and count the number of providers in each county
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# read csv file
df = pd.read_csv("COVID-19 in Alabama Vaccine Providers-2.csv")
county_count = df.groupby('County')['Name'].count().sort_values(ascending = False)
print(county_count)
plt.figure(figsize=(5, 15))
sns.set_color_codes("pastel")
sns.barplot(y = county_count.index, x = county_count.values, orient = "h", color="b")
plt.title('Number of Providers (descending order) ')
plt.xlabel('Number of Providers By County')
plt.ylabel('County')
plt.show()
```

```python
#3 Covid-19 Vaccine Data: calculate the rate of people who are completely vaccinated in all counties in AL, and sort these counties by the complete vaccinated rate in descending order.
import requests
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
df = pd.read_csv("COVID-19 County Vaccine Data-2.csv")
indexAge = df[ (df['Population > 16'] < 0)  ].index
df.drop(indexAge , inplace=True)
# change data to numeric type
df['People_Completely_Vaccinated'] = pd.to_numeric(df['People Completely Vaccinated'], errors='coerce')
df['Population'] = pd.to_numeric(df['Population > 16'], errors='coerce')

People_Completely_Vaccinated = df.groupby('County')['People Completely Vaccinated'].sum()
Population = df.groupby('County')['Population'].sum()
complete_vaccinated_rate = (People_Completely_Vaccinated/Population).sort_values(ascending = False)
print(complete_vaccinated_rate)
plt.figure(figsize=(5, 15))
sns.set_color_codes("pastel")
sns.barplot(y = complete_vaccinated_rate.index, x = complete_vaccinated_rate.values, orient = "h", color="b")
plt.title('Complete Vaccinated Rate (descending order) ')
plt.xlabel('Complete Vaccinated Rate by County')
plt.ylabel('County')
plt.show()
```
