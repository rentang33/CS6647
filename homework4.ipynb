```python
#Question #1
#import libraries
import matplotlib.pyplot as plt
import numpy as np
import random
#create arries to store the probability valies
switch_probability = []
stick_probability =[]
boxes =["ten","ten",'hundred']

#Monte Carlo Simulation, N times
def monte_carlo(N):
  #counting wins by swith and stick
  win_switch_counter =0
  win_stick_counter =0
  for i in range(N):
    # randomize the order
    random.shuffle(boxes)
    # intial choice from 3 box
    k= random.randrange(0,2)
    if boxes[k] !='hundred':
      win_switch_counter +=1
    else:
        win_stick_counter +=1
    #Calculation probabilites
    win_by_switch = win_switch_counter/(i+1)
    win_by_stick = win_stick_counter/(i+1)

    #Updateing the list values
    switch_probability.append(win_by_switch)
    stick_probability.append(win_by_stick)
  return switch_probability[-1],stick_probability[-1]

#Callthe function and run 1000 of N
N =1000
switch, stick = monte_carlo(N)
#Print answer
print("The answer is to switch, because if I don't switch, I have a 1/3 chance of getting the $100 bill. \nSince my friend shows me another box with a $10 bill, so if I switch, I will have a 2/3 chance of getting it, \nso switching gives me a better chance of getting the $100 bill.")
print ("\nAfter runing", N, "tiemes, the probability is: ")
print("Probability of getting $100 by switching: ", switch)
print("Probability of getting $100 by sticking: ", stick)
#Plotting the data
plt.plot(switch_probability)
plt.plot(stick_probability)
plt.title  ("Monte Carlo method for guessing $100 bill")
plt.xlabel("Number of runs.    (Blue = Switching, Orange = Sticking)")
plt.ylabel("Probability")

#Reference
#Saleem, A. (2022, October 22). Solving the Monty Hall problem with Monte Carlo simulation in Python. Data Science Dojo. https://datasciencedojo.com/blog/monte-carlo-simulation-python/
```
